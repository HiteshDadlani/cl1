```python
df_json = pd.read_json("StoreSales (1).json", encoding="latin1")#this command helps
#to read the json file and here encoding="latin1" is used because if we dont use this then an 
#error will be there because when a text file .json is saved on Windows system and if we want 
#to read all the content or we can say the characters including the
#special characters we use this encoding which helps Python to read whole content
df_json
```

```python
# Fill object columns (text) with mode and numeric columns with mean
for col in df_json.columns:#visiting each column of the dataset 
    if df_json[col].isnull().any():#We are first finding out the columns which have null values in the dataset
        if df_json[col].dtype == 'object':#Now we are segregating the object data type columns on one side
            df_json[col].fillna(df_json[col].mode()[0], inplace=True)#We are filling the missing values of object data type columns
            #using the mode of that particular column using fillna method
        else:
            df_json[col].fillna(df_json[col].mean(), inplace=True)#We are filling the missing values of numeric data type columns
            #or we can say columns having datatype other than object data type using the mean of that particular column using fillna method
# Checking if any missing values remain
df_json.isnull().sum()
```

```python
df=df.drop_duplicates()#We are dropping duplicate rows from csv file
df_excel = df_excel.drop_duplicates()#We are dropping duplicate rows from excel file
df_json = df_json.drop_duplicates()#We are dropping duplicate rows from json file
```

```python
df['Source'] = 'CSV'#This adds a new column in csv dataset named "Source" which will tell that from
#where that particular row record has come 
df_excel['Source'] = 'Excel'#This adds a new column in excel dataset named "Source" which will tell that from
#where that particular row record has come
df_json['Source'] = 'JSON'#This adds a new column in json dataset named "Source" which will tell that from
#where that particular row record has come
```

```python
unified_df = pd.concat([df, df_excel, df_json], ignore_index=True)#We merged
#all CSV,JSON and Excel Dataset in one new Dataset named unified_df using pd.concat command
#ignore_index=True this gives the row numbers indexing from 0,1,2,3,4.........
#If we do not did ignore_index=True then every for every first value of different dataset the row would have 
#been starting from zero means suppose lets suppose thats first four values in unified_df are from
#csv file then row numbering would have been 0,1,2,3 and then suppose the fifth record in unified_df is from
#Excel dataset so again the row numbering would have been started from 0 so the numbering would have been 
#0,1,2,3,0 thats not correct thats why we used this
```

```python
print("Unified dataframe shape:", unified_df.shape)#unified_df.shape tells how many rows and columns are there inthe merged 
#dataset
unified_df.head()#Printing first 5 values of unified_df dataset
```

```python
unified_df['Order Date'] = pd.to_datetime(unified_df['Order Date'], errors='coerce')
#It converts the “Order Date” column in your dataframe
#from text (string) format → into date/time (datetime) format.

#What errors='coerce' means
#If some rows contain invalid dates (e.g., “N/A”, “13/35/2023”),
#instead of crashing with an error, Pandas will replace them with NaT (missing datetime).
#So this makes your code safe and clean.
unified_df['Year'] = unified_df['Order Date'].dt.year#extracts the year from Order Date column and make a new column named "Year"in unified_df
unified_df['Month'] = unified_df['Order Date'].dt.month#extracts the month from Order Date column and make a new column named "month"in unified_df
unified_df['Day'] = unified_df['Order Date'].dt.day#extracts the day from Order Date column and make a new column named "day"in unified_df
```

```python
unified_df['Amount'] = unified_df['Quantity'] * unified_df['Unit price']#It makes a new column named Amount in unified dataset and stores the calculated amount 
unified_df['Profit Margin %'] = (unified_df['Profit'] / unified_df['Sales']) * 100#It makes a new column named ProfitMargin % in unified dataset and stores the calculated Profit Margin %
```

```python
import matplotlib.pyplot as plt
unified_df.groupby('Category')['Sales'].sum().plot(kind='bar', color='teal')#Groups the dataset by Category (like Furniture, Technology, Office Supplies)
#and  Calculates the total Sales for each category.
#and it Uses the teal color for bars.
plt.title("Total Sales by Category")#Title of Bar Plot
plt.xlabel("Category")#Name on X axis
plt.ylabel("Sales")#Name on Y axis
plt.show()#Displays the plot
```

```python
unified_df['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)#Counts how many times each product category appears in your dataset.
#autopct='%1.1f%% means keep only 1 digit after decimal
#startangle=90
#→ Rotates the pie chart start position by 90° (makes it look nicer).
plt.title("Product Category Distribution")
plt.ylabel("")
plt.show()#Displays the Pie Chart
```

```python
unified_df.groupby('Region')['Sales'].sum().plot(kind='bar', color='orange')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```
