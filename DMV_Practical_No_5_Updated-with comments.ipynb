```python
df_aqi = df[df['AQI'] > 0]#Ensures all AQI values are greater than zero because AQI cannot be negative
```

```python
 df['Date'] = pd.to_datetime(df['Date'])#converted Date column to datetime format
```

```python
df=df.drop_duplicates()#removes duplicate records
```

```python
df_aqi = df[df['AQI'] > 0].copy()
df_aqi['Date'] = pd.to_datetime(df_aqi['Date'])   # convert to datetime

# Group by month
monthly_aqi = df_aqi.set_index('Date')['AQI'].resample('M').mean().reset_index()
#This groups data by month using the date as an index.
#resample('M') → means “group by month.”
#Then .mean() → takes the average AQI for each month.
#Result → one AQI value per month (smooth trend).
# Plot
plt.figure(figsize=(10,6))
plt.plot(monthly_aqi['Date'], monthly_aqi['AQI'], color='blue', linewidth=2, marker='o')
#markers='o' draw the blue points/circles on the line plot
#linewidth=2 controls the thickness of the blue line in the line plot
plt.title('Average Yearly AQI Trend')
plt.xlabel('Month')
plt.ylabel('Air Quality Index (AQI)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

#The blue dots in between the lines represents the Months from Jan to Dec only in 2015 the Month 
#Starts from Feb because from 2015 to 2016 there are only 11 dots instead of 12 and 2020 has 
#only 8 months because 8 dots are there for 2020 you can count the dot
```

```python
 plt.figure(figsize=(10,6))
 # Plot for PM2.5
 plt.subplot(3, 1, 1)
#This line is used to create multiple plots (subplots) in a single figure
#Syntax is:-
#plt.subplot(nrows, ncols, index)
#nrows=3 means The figure will have 3 rows of plots.
#ncols means the figure will have only one column means all three plots will be arranged vertically one below
#to other
#The last 1 represents that first plot among the three 
 plt.plot(df['Date'], df['PM2.5'], color='red', label='PM2.5')

 plt.title('PM2.5 Trend Over Time')
 plt.ylabel('PM2.5 Levels')
 plt.grid(True)
 # Plot for PM10
 plt.subplot(3, 1, 2)#the last two represents middle plot basically its the plot number
 plt.plot(df['Date'], df['PM10'], color='green', label='PM10')
 plt.title('PM10 Trend Over Time')
 plt.ylabel('PM10 Levels')
 plt.grid(True)
 # Plot for CO
 plt.subplot(3, 1, 3)
 plt.plot(df['Date'], df['CO'], color='purple', label='CO')
 plt.title('CO Trend Over Time')
 plt.ylabel('CO Levels')
 plt.xlabel('Date')
 plt.grid(True)
 plt.tight_layout()
 plt.show()

#PM2.5 levels remain high and show repeated spikes over time, suggesting consistent 
#fine-particle pollution possibly from traffic, construction, and burning activities.There are multiple sharp spikes, especially around 2016, 2018, and 2019 — 
#indicating periods of very poor air quality.

#Around 2018–2019, PM10 values rise significantly, 
#Showing a period of heavier dust or coarse particulate pollution.

#CO (carbon monoxide) levels are much lower overall compared to particulate matter.

#However, small spikes can be seen around 2017–2019, 
#possibly linked to vehicle emissions or burning events.
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['Date'], df['AQI'], color='skyblue')
plt.title('AQI Comparison Across Different Time Periods')
plt.xlabel('Date')
plt.ylabel('Air Quality Index (AQI)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(x='AQI_Bucket', y='AQI', data=df, palette='coolwarm')
plt.title('AQI Distribution Across Categories')
plt.xlabel('AQI Category')
plt.ylabel('Air Quality Index (AQI)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(df['PM2.5'], df['AQI'], color='red', label='PM2.5')
plt.scatter(df['PM10'], df['AQI'], color='green', label='PM10')
plt.scatter(df['CO'], df['AQI'], color='purple', label='CO')

plt.title('Relationship Between AQI and Pollutant Levels')
plt.xlabel('Pollutant Levels')
plt.ylabel('Air Quality Index (AQI)')
plt.legend()#Craetes the legend
plt.grid(True)#Creates grid lines
plt.show()
#PM2.5 points rise steeply with AQI.
#The higher the PM2.5 value, the higher the AQI tends to be.
#This shows a strong positive correlation between PM2.5 and AQI.

#PM10 also shows an upward pattern but more spread out.
#At lower AQI values (below 500), many PM10 points are clustered, indicating moderate influence on AQI.

#CO values stay near the left (lower x-axis), showing much smaller concentrations.
#However, some CO spikes still correspond to higher AQI values.
```
